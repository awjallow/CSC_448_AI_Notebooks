# Abstract
The program in this notebook is an implementation of a Binary Search Tree (BST) that I wrote in Python when I was 
practicing for technical interviews related to Data Structures & Algorithms. The BST class has methods that perform the most common operations on a BST. I wrote a test function to test out the BST class implementation and its methods.

```python
# Node class
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


# Binary Search Tree(BST) class
class BST:
    def __init__(self):
        self.root = None
        self.depth = 0

    def insert(self, root, new_node):
        if root is None:
            root = new_node
        else:
            if new_node.data < root.data:
                if root.left == None:
                    root.left = new_node
                else:
                    self.left = self.insert(root.left, new_node)
            else:
                if root.right == None:
                    root.right = new_node
                else:
                    self.insert(root.right, new_node)

    def print_in_order(self, root, values):
        if root is None:
            return

        self.print_in_order(root.left, values)
        values.append(root.data)
        self.print_in_order(root.right, values)

        return values

    def print_pre_order(self, root, values):
        if root:
            values.append(root.data)
            self.print_pre_order(root.left, values)
            self.print_pre_order(root.right, values)

        return values

    def print_post_order(self, root, values):
        if root is None:
            return

        self.print_post_order(root.left, values)
        self.print_post_order(root.right, values)
        values.append(root.data)

        return values

    def print_level_order(self, root, levels):
        if not root:
            return levels

        queue = [root]

        while queue:
            level = []
            queue_len = len(queue)

            for i in range(queue_len):
                curr_node = queue.pop(0)
                level.append(curr_node.data)

                if curr_node.left:
                    queue.append(curr_node.left)
                if curr_node.right:
                    queue.append(curr_node.right)

            levels.append(level)

        self.depth = len(levels)

        return levels

    def find_min(self, root):
        if root.left is None:
            return root.data

        return self.find_min(root.left)

    def find_max(self, root):
        if root is None:
            return

        if root.right is None:
            return root.data

        return self.find_max(root.right)

    def find_depth(self, root):
        if not root:
            return 0

        return self.depth

    def contains(self, root, val):
        if root is None:
            return

        if root.data == val:
            return True
        elif val < root.data:
            if root.left is None:
                return False
            else:
                return self.contains(root.left, val)
        else:
            if root.right is None:
                return False
            else:
                return self.contains(root.right, val)

    # incomplete
    def delete_node(self, val):
        if root is None:
            return


'''
          10
         /  \
        5    12
       / \   / \
    -10   6 11  17
                  \
                  77

    [[10], [5, 12], [-10, 6, 11, 17], [77]]

'''

bst = BST()
root = Node(10)
bst.insert(None, root)
bst.insert(root, Node(5))
bst.insert(root, Node(-10))
bst.insert(root, Node(6))
bst.insert(root, Node(12))
bst.insert(root, Node(11))
bst.insert(root, Node(17))
bst.insert(root, Node(77))


def test():
    # print(bst.print_in_order(root, []))
    in_order = [-10, 5, 6, 10, 11, 12, 17, 77]
    print(in_order == bst.print_in_order(root, []))  # expects True

    # print(bst.print_pre_order(root, []))
    pre_order = [10, 5, -10, 6, 12, 11, 17, 77]
    print(pre_order == bst.print_pre_order(root, []))  # expects True

    # print(bst.print_post_order(root, []))
    post_order = [-10, 6, 5, 11, 77, 17, 12, 10]
    print(post_order == bst.print_post_order(root, []))  # expects True

    min_val = -10
    print(min_val == bst.find_min(root))  # expects True

    max_val = 77
    print(max_val == bst.find_max(root))  # expects True

    level_order = [[10], [5, 12], [-10, 6, 11, 17], [77]]
    print(level_order == bst.print_level_order(root, []))  # expects True

    depth = 4
    print(depth == bst.find_depth(root))

    target = 6
    print(bst.contains(root, target) is True)

    return "\nAll tests executed!"


print(test())
```
